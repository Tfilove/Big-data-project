## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.


df_clean = df.copy()

# Handle Missing Values

# Convert Rating to numeric
df_clean['Rating'] = pd.to_numeric(df_clean['Rating'], errors='coerce')

# Fill missing Rating values with median
df_clean['Rating'].fillna(df_clean['Rating'].median(), inplace=True)

# Remove Duplicate Records
df_clean.drop_duplicates(inplace=True)

# Data Type Correction

# Convert Date column to datetime
df_clean['Date'] = pd.to_datetime(df_clean['Date'])

# Convert numerical columns to numeric
df_clean['Unit price'] = pd.to_numeric(df_clean['Unit price'], errors='coerce')
df_clean['Quantity'] = pd.to_numeric(df_clean['Quantity'], errors='coerce')

# Feature Engineering

# Create Total Sales feature
df_clean['Total_Sales'] = df_clean['Unit price'] * df_clean['Quantity']

# Final Check
df_clean.info()

```

### What all manipulations have you done and insights you found?

During the data wrangling stage, several preprocessing steps were carried out to make the supermarket sales dataset suitable for analysis. Initially, the dataset structure was examined using the info() function to understand the number of records, columns, and data types. Missing values were identified, and it was observed that the Rating column contained missing entries, which were handled by converting the column to numeric format and filling the missing values using the median to maintain data consistency. Duplicate records were also checked and removed to ensure accuracy. Data types were verified and corrected, particularly for numerical columns such as Unit price and Quantity, to enable proper calculations. Additionally, a new feature called Total_Sales was created by multiplying unit price with quantity, which enhanced analytical insights related to sales performance. These preprocessing and feature engineering steps ensured that the dataset was clean, consistent, and ready for effective exploratory data analysis.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
# Chart - 1 visualization code

#  Distribution of Customer Ratings
plt.figure(figsize=(8, 5))
plt.hist(df_clean['Rating'], bins=10)
plt.title("Distribution of Customer Ratings - Supermarket Sales")
plt.xlabel("Rating")
plt.ylabel("Number of Customers")
plt.show()

```

##### 1. What is/are the insight(s) found from the chart?

Most customer ratings are concentrated between 6 and 8, indicating generally positive customer satisfaction with the supermarket services. Extremely low or very high ratings are comparatively fewer, suggesting consistent service quality across customers.

#### Chart - 2

```python
# Chart - 2 visualization code

# Total Sales by Product Line
product_sales = df_clean.groupby('Product line')['Total_Sales'].sum()
plt.figure(figsize=(8, 5))
product_sales.plot(kind='bar')
plt.title("Total Sales by Product Line - Supermarket Sales")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

The chart indicates that the Lifestyle product line generates the highest total sales, highlighting strong customer demand in this category. In contrast, Health & Accessories records the lowest total sales, suggesting comparatively lower customer preference or sales volume among the product lines.

#### Chart - 3

```python
# Chart - 3 visualization code

# Branch-wise Total Sales
branch_sales = df_clean.groupby('Branch')['Total_Sales'].sum()
plt.figure(figsize=(8, 5))
branch_sales.plot(kind='bar')
plt.title("Branch-wise Total Sales - Supermarket Sales")
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

The chart shows that Brooklyn records the highest total sales among the branches, indicating slightly better sales performance compared to the others. However, total sales across Manhattan and Queens are also very close, suggesting that sales performance is relatively consistent across all branches.

#### Chart - 4

```python
# Chart - 4 Pie Chart

# Payment Method Distribution
payment_counts = df_clean['Payment'].value_counts()
plt.figure(figsize=(8, 5))
payment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Payment Method Distribution - Supermarket Sales")
plt.ylabel("")  # Hide y-axis label for clean look
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

The chart indicates that E-wallets are the most preferred payment method, accounting for the largest share of transactions, which reflects a growing adoption of digital payments. Cash and credit card payments are also widely used, showing that customers continue to rely on multiple payment options while shopping at the supermarket.

#### Chart - 5

```python
# Chart - 5 visualization code

# Total Sales by Customer Type
customer_sales = df_clean.groupby('Customer type')['Total_Sales'].sum()
plt.figure(figsize=(8, 5))
customer_sales.plot(kind='bar')
plt.title("Total Sales by Customer Type - Supermarket Sales")
plt.xlabel("Customer Type")
plt.ylabel("Total Sales")
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

The chart shows that Normal customers contribute higher total sales compared to Member customers, indicating that non-members make more purchases overall. However, members still represent a significant share of sales, suggesting potential for increasing revenue through stronger membership engagement and loyalty programs.

#### Chart - 6 - Correlation Heatmap

```python
# Correlation Heatmap visualization code

# Select numerical columns
numerical_data = df_clean[['Unit price', 'Quantity', 'Rating', 'Total_Sales']]

# Compute correlation matrix
corr_matrix = numerical_data.corr()

# Plot correlation heatmap
plt.figure(figsize=(8, 5))
plt.imshow(corr_matrix)
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title("Correlation Heatmap - Supermarket Sales")
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

The heatmap shows a strong positive correlation between Total_Sales and both Unit price and Quantity, indicating that higher prices and larger quantities directly increase total sales. In contrast, Rating has a weak or negligible correlation with the other variables, suggesting that customer ratings do not significantly influence sales value in this dataset.

#### Chart - 7 - Pair Plot

```python
# Pair Plot visualization code

# Select numerical columns for pair plot
pairplot_data = df_clean[['Unit price', 'Quantity', 'Rating', 'Total_Sales']]

# Create pair plot
sns.pairplot(pairplot_data)

plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

1. There is a visible positive relationship between Unit price and Total_Sales, indicating that higher-priced items generally contribute more to overall sales.
2. Quantity and Total_Sales show a strong positive relationship, meaning purchases in larger quantities significantly increase sales value.
3. Ratings do not display a strong relationship with Unit price or Quantity, suggesting that customer ratings are not directly influenced by price or purchase volume.
4. Ratings are mostly concentrated around the mid to higher range, reflecting generally positive customer satisfaction across transactions.
